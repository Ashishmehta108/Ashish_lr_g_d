```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
class Ashish_lr:
     def __init__(self,epoch,learning_rate):
         self.m=0
         self.b=0

         self.epoch=epoch
         self.learning_rate=learning_rate

     def fit(self,X,y):



         for i in range (0,self.epoch):

             loss_gr_m=2*np.sum((np.ravel(y)-self.m*np.ravel(X)-self.b)*(-np.ravel(X)))
             loss_gr_b=2*np.sum(((np.ravel(y)-self.m*np.ravel(X)-self.b)*(-1)))
             self.m=self.m-self.learning_rate*(loss_gr_m)
             self.b=self.b-self.learning_rate*(loss_gr_b)

         return ({"intercept from Ashish model":b,"slope from Ashish model":m})
```

```python
df=pd.read_csv("Now.csv")
df.head(4)
```

```python
X=df.iloc[:,0:1]
Y=df.iloc[:,1:]
X_train,X_test,y_train,y_test=train_test_split(X,Y,test_size=0.33,random_state=2)
```

```python
Lr=Ashish_lr(425,0.01)
```

#for this data epoch=425 and learning rate=0.01 is best

```python
Lr.fit(X_train,y_train)
```
